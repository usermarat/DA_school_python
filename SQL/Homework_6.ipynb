-- Далее построить гистаграмму распределения этих трех колонко

```python
import pandas as pd
import numpy as np
from random import randint
import sqlite3
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
conn = sqlite3.connect('task1_7.db')
pd.DataFrame([[randint(0,1000) for i in range(3)]
              for j in range(1000)]).to_sql('table1', conn)
```

```python
df = pd.read_sql_query('select * from table1', conn, index_col='index')
df.hist(sharex=True, sharey=True, figsize=(15,7));
```

join Employee b\
on b.Id = a.ManagerId\
where a.Salary > b.Salary\
and a.ManagerId is not null
